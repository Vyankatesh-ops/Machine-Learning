```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe=OneHotEncoder()
```

```python
df_array=ohe.fit_transform(df[['Company','Place','Gender']]).toarray()
df_array
```

```python
single_list=[innerlistone for main_array in ohe.categories_ for innerlistone in main_array]
single_list
```

```python
df_new=pd.DataFrame(df_array,columns=single_list)
df_new
```
